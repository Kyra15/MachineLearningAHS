```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read csv
df = pd.read_csv("data/titanic.csv")
df.head()
```

```python
df.isna().any()
```

From this table, we can tell that the age, cabin, and embarked columns all have NaN values, so we'll have to clean up those, mostly likely by dropping the row altogether (mean imputation might cause biases due to gender).

```python
print(f"Maximum age: {max(df['Age'])}")
print(f"Minimum age: {min(df['Age'])}")
print(f"DF of passengers whose ages are less than 1:\n {df[df['Age'] < 1]}")
```

The maximum age is 80, which is high and rare for this time, but not entirely out of the question since these passengers were mostly wealthy with easy access to medical help. The minimum age, on the other hand, is 0.42. This is a little weird since age isn't usually represented with decimals, however it might just be a baby who's not yet 1. Upon further analysis, it seems that many passengers had ages less than 1, with one paying a fare of 150 dollars. This now seems more like an error than actual fact.

From these current columns, I would think that a few wouldn't be helpful to our overall predictions. Ticket number, passengerid, sibsp (siblings/spouses onboard), parch (number of parents/children on board), embarked (port of embarkation), name, and cabin are the attributes I believe won't contribute as much towards out goal. Even if they make small differences (having an earlier ticket implies more wealth), I think they'll be neglible enough to exclude them for training purposes. 

label descriptions taken from: https://github.com/awesomedata/awesome-public-datasets/issues/351

Additionally, there are a few columns that need to be one-hot-encoded, like sex (0 for female, 1 for male). The fare and age might also need to be normalized. 

```python
# graph age and survival rate
women = df[df["Sex"] == "female"][["Age", "Survived"]].dropna()
men = df[df["Sex"] == "male"][["Age", "Survived"]].dropna()

plt.hist([women["Age"], men["Age"]], bins=10, color=["Red", "Blue"], label=["Women", "Men"])
plt.title("Number of people on the Titanic vs. Age Separated by Gender")
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.legend()
plt.show()

# get survival rates based on gender
w_survival = women["Survived"].mean()
print(f"Women survival rate: {w_survival}")

m_survival = men["Survived"].mean()
print(f"Men survival rate: {m_survival}")
```

```python
# graph a histogram of ages, with gender separation, but show survival rates for each bin

# basically group the women into the bin categories, basically based on tens round(-1), then find the mean survivals of those

women["Age"] = round(women["Age"], -1)

grouped_women = women.groupby(women["Age"])

women_avgs = grouped_women["Survived"].mean()

women_avgs[70.0] = 0
women_avgs[80.0] = 0

print(women_avgs)


men["Age"] = round(men["Age"], -1)

grouped_men = men.groupby(men["Age"])

men_avgs = grouped_men["Survived"].mean()

print(men_avgs)

survival_df =  pd.DataFrame({"Women": women_avgs, "Men": men_avgs})

survival_df.plot(kind="bar", width=0.8, color=["red", "blue"])
plt.ylabel("Survival Rate")
plt.title("Age vs. Survival Rate based on Gender")
plt.show()

total_df = (survival_df["Women"] * len(survival_df["Women"]) + survival_df["Men"] * len(survival_df["Men"])) / (len(survival_df["Men"]) + len(survival_df["Women"]))
print(total_df)

total_df.plot(kind="bar", width=0.8)
plt.title("Age vs. Survival Rate")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# relationship between fare prices and class on ship
fare_pclass_df = df[["Fare", "Pclass"]]

plt.scatter(fare_pclass_df["Pclass"], fare_pclass_df["Fare"])
plt.xticks([1, 2, 3])
plt.xlabel("Passenger Class")
plt.ylabel("Fare Cost")
plt.title("Passenger Class vs. Fare Cost")
plt.show()

grouped_fare = fare_pclass_df.groupby(df["Pclass"])
print(grouped_fare.mean())

plt.plot(grouped_fare.mean()["Fare"], marker=".", markersize=10)
plt.show()
```

I also noticed that there was a singular ticket price that was $500, which is a clear outlier.

```python
fare_pclass_df_noout = df[["Fare", "Pclass"]]
fare_pclass_df_noout = fare_pclass_df[fare_pclass_df["Fare"] < 400]

plt.scatter(fare_pclass_df_noout["Pclass"], fare_pclass_df_noout["Fare"])
plt.xticks([1, 2, 3])
plt.xlabel("Passenger Class")
plt.ylabel("Fare Cost")
plt.title("Passenger Class vs. Fare Cost")
plt.show()

grouped_fare_noout = fare_pclass_df.groupby(df["Pclass"])
print(grouped_fare_noout.mean())

plt.plot(grouped_fare_noout.mean()["Fare"], marker=".", markersize=10)
plt.xticks([1, 2, 3])
plt.xlabel("Passenger Class")
plt.ylabel("Mean Fare Cost")
plt.title("Passenger Class vs. Mean Fare Cost")
plt.show()
```

Even with removing that outlier, the data is still pretty skewed, indicating a large class divide between 1 and 2/3.

```python
classes_df = df[["Survived", "Pclass"]]
grouped_class = classes_df.groupby(df["Pclass"])

class_avgs = grouped_class.mean()

print(class_avgs)

class_avgs["Survived"].plot(kind="bar", ylim=(0, 1.0))
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Passenger Class vs. Survival Rates")
plt.show()
```

# 4.Prepare the Data


Apply any data transformations and explain what and why


```python
# grab cols i want
model_df = df[["Survived", "Age", "Sex", "Pclass", "Fare"]]
model_df.head()
```

These classes showed the strongest correlations for survival rate, and they were the easiest to manipulate so I've saved only them for the dataframe.

```python
# do data transformations like one hot encoding and making bins for fare prices
one_hot_model_df = pd.get_dummies(model_df, columns=["Sex"], dtype='int').dropna()
one_hot_model_df.rename(columns={"Sex_female": "Is_female"}, inplace=True)
one_hot_model_df['FareBin'] = pd.qcut(one_hot_model_df['Fare'], 4, labels=False)
fare_bins = pd.qcut(one_hot_model_df['Fare'], 4, labels=False, retbins=True)[1]
one_hot_model_df.drop(columns=["Sex_male", "Fare"], inplace=True)

one_hot_model_df.head()
```

```python
# import scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

# 5. Model the data
Using selected ML models, experment with your choices and describe your findings. Finish by selecting a Model to continue with


```python
# import libraries
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
# separate into data + labels
features = ['Age', 'Pclass', 'Is_female', 'FareBin']

X = one_hot_model_df[features]
y = one_hot_model_df['Survived']
target_names = ["Did not survive", "Survived"]
```

```python
# splt into train test
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.25, 
                                                    random_state=42, 
                                                    stratify=y)

scaler.fit(X_train)

# scale x
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# define decision tree + fit
dtree = DecisionTreeClassifier(random_state=42)

dtree = dtree.fit(X_train, y_train)
```

```python
# do a prediction
y_pred_tree = dtree.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_tree)
print(f'Model Accuracy: {accuracy}')

print("\nClassifiation Report:")
print(classification_report(y_test, y_pred_tree, target_names=target_names))
```

```python
# define SVM one v one + fit
svm_ovo = SVC(decision_function_shape='ovo')
svm_ovo.fit(X_train, y_train)
```

```python
# do a prediction
y_pred_ovo = svm_ovo.predict(X_test)
print("One-vs-One Accuracy:", accuracy_score(y_test, y_pred_ovo))

print("\nClassifiation Report:")
print(classification_report(y_test, y_pred_ovo, target_names=target_names))
```

```python
# define random forest + fit
from sklearn.ensemble import RandomForestClassifier

rf_class = RandomForestClassifier(n_estimators=100, random_state=42)

rf_class.fit(X_train, y_train)
```

```python
# do a prediction
y_pred_rf = rf_class.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))

print("\nClassifiation Report:")
print(classification_report(y_test, y_pred_rf, target_names=target_names))
```

```python
# define logisitc regression + fit
log = LogisticRegression(random_state=42, max_iter=1000)
log.fit(X_train_scaled, y_train)
```

```python
# do a prediction
y_pred_log = log.predict(X_test_scaled)
print("Logistic Accuracy:", accuracy_score(y_test, y_pred_log))

print("\nClassifiation Report:")
print(classification_report(y_test, y_pred_log, target_names=target_names))
```

# 6. Fine Tune the Model

With the select model descibe the steps taken to acheve the best rusults possiable 


```python
# ok so model accuracy is not very good lets train some hyper parameters!

# https://www.geeksforgeeks.org/machine-learning/how-to-tune-a-decision-tree-in-hyperparameter-tuning/
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from scipy.stats import randint
```

```python
# got a bunch of irrelavent warnings while tuning logisitc so i got rid of them
import warnings
from sklearn.exceptions import FitFailedWarning

warnings.filterwarnings("ignore", category=FitFailedWarning)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
# def parameters to change and random search through them for best log model
param_grid = [{'penalty':['l2', 'None'],
               'C' : np.logspace(-4,4,20),
               'solver': ['lbfgs','newton-cg','liblinear','sag','saga'],
               'tol': [1e-4, 1e-3, 1e-2]
             }]

log_2 = LogisticRegression(random_state=42, max_iter=1000)

rand_log = RandomizedSearchCV(log_2, param_distributions=param_grid, 
                           cv=5, n_iter=100, random_state=42)
rand_log.fit(X_train_scaled, y_train)

print("Best Params for random for log:", rand_log.best_params_)
print("Best CV Score for random for log:", rand_log.best_score_)
best_log = rand_log.best_estimator_
```

```python
# def parameters to change and random search through them for best random forest model
param_dist = {'n_estimators': randint(50,500),
              'max_depth': [None] + list(range(5, 51, 5)), 
              'min_samples_split': randint(2, 20),
              'bootstrap': [True, False]
             }

rf_class2 = RandomForestClassifier(random_state=42, max_samples=0.5)

rand_rf = RandomizedSearchCV(rf_class2, 
                                 param_distributions = param_dist, 
                                 n_iter=50, 
                                 cv=5, 
                                 random_state=42)

rand_rf.fit(X_train, y_train)
print("Best Params for random for random forest:", rand_rf.best_params_)
print("Best CV Score for random for random forest:", rand_rf.best_score_)
best_rf = rand_rf.best_estimator_
```

# 8. Launch the Model System
Define your production run code, This should be self susficent and require only your model pramaters 


```python
def infrence(prams):

    # take input (wasn't sure how data would be formatted)
    input_df = pd.DataFrame([prams], columns=features)
    input_df["Is_female"] = 1 if prams[2] == "female" else 0
    input_df["FareBin"] = pd.cut([prams[3]], bins=fare_bins, labels=False, include_lowest=True)[0]

    # scale
    new_X = input_df
    new_X_scaled = scaler.transform(new_X)

    # make predictions
    log_pred = best_log.predict(new_X_scaled)
    log_probs = best_log.predict_proba(new_X_scaled)
    rf_pred = best_rf.predict(new_X)
    rf_probs = best_rf.predict_proba(new_X)

    # print predictions
    print(f"Survived?: {'Yes' if log_pred[0] == 1 else 'No'}. Probability of survival: {round(log_probs[0][1]*100, 2)}% (based on logistic model)")
    print(f"Survived?: {'Yes' if rf_pred[0] == 1 else 'No'}. Probability of survival: {round(rf_probs[0][1]*100, 2)}% (based on random forest model)")

    # return results
    results = {"Log": [log_pred, log_probs],
              "RF": [rf_pred, rf_probs]}
    return results 
```

```python
# test case with sa passenger 29 years old, first class, female, and paid $75
infrence([29, 1, "female", 75])
```
